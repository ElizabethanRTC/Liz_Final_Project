```python
import pandas as pd
import numpy as np
```

```python
path="2021_Substack_Survey_Clean.csv"
df = pd.read_csv(path)
df
```

```python
q1 = np.quantile(df["ConvertedCompYearly"], .25)
q3 = np.quantile(df["ConvertedCompYearly"], .75)
IQR = q3-q1
cut_high = (1.5*IQR)+q3
cut_low = q1-(1.5*IQR)
print(f" Q1: {q1},Q3: {q3}, IQR: {IQR}, outlier high: {cut_high}, outlier low: {cut_low}")
```

```python
def new_column(row):
    if row["ConvertedCompYearly"] <= q1:
        return 1
    elif row["ConvertedCompYearly"] >= cut_high:
        return 1
    else:
        return 0
    
df["outlier"] = df.apply(new_column, axis=1)
df
```
